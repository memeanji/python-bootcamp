### **loc, iloc 사용하기**

- loc: 레이블값을 사용하여 조회
 - `데이터프레임명.loc[행조건,열조건]`
 - 열만 조회할때는 행조건에 `:`를 입력합니다.

```python
df.loc[3,]
```

```python
df.loc[:,'Name'] #name이라는 이름을 가진 열이 출력
```

```python
df.loc[3,'Name'] # 인덱스가 3인 행의 이름 컬럼의 값이 불러와짐
```

```python
df.loc[3:5,] # 3에서 5까지 행 전체를 가져옴
```

```python
df.loc[3:5,['Age','Pclass']] #3에서 5까지의 행과 age와 pclass의 열이 불러와짐
```

```python
df.loc[[1,3,5],['Age','Pclass']] # 특정 인덱스르 가져올 수도 있음
```

## **[2] 원하는 조건으로 데이터 추출하기**

### **데이터 정렬하기**
- `데이터프레임명.sort_values(정렬기준컬럼)`
- 내림차순으로 정렬하고 싶다면 `ascending=False` 조건을 걸어주면 됩니다.

```python
df.head()
```

```python
df.sort_values('Age')  #Age 컬럼을 기준으로 정렬하고 싶어
```

```python
df.sort_values('Age', ascending=False) # 내림차순 정렬
```

```python
df.sort_values(['Age','Fare'], ascending=[False,True]) # Age는 내림 Fare는 오름
```

### **특정 조건을 충족하는 데이터 추출하기**
- `데이터프레임명[조건식]`
- `데이터프레임명.query('조건식')`



```python
df[df['Pclass'] == 1] # pclass가 1인 것만 가져올 수 있음
```

```python
df[(df['Pclass'] == 1) & (df['Age'] >= 30)]  #pclass가 1이고 age가 30 이상인걸 들고와줘
```

```python
df[(df['Pclass'] == 1) | (df['Age'] >= 30)] #pclass가 1이거나 age가 30 이상인걸 들고와줘
```

```python
df.query('Pclass == 1')
#쿼리함수 안에선 문자열 안에 들어가야함
```

```python
df.query('Pclass == 1 and Age >= 30')
```

```python
df.query('Pclass == 1 or Age >= 30')
```

```python
df[df['PassengerId'].isin([3,100,500])] #passengerId가 3,100,500인것만 들고옴
```

```python
df.query('PassengerId in [3,100,500]')
```

```python
df.query('PassengerId.isin([3,100,500])')
```

```python
import random

passengerid_list = list(df['PassengerId'])
passengerid_sample = random.sample(passengerid_list, 10)
```

```python
df[df['PassengerId'].isin(passengerid_sample)]
```

```python
df.query('PassengerId in @passengerid_sample')
#@를 붙이면 해당 리스트를 참조해서 조회함
```
