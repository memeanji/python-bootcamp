### **인덱스 변경**
- 인덱스 몇개를 바꿀 때
    - `데이터명.rename({인덱스:바꿀 인덱스, 인덱스:바꿀 인덱스, ...})`
- 인덱스 전체를 바꿀 때
    - `데이터명.index = 바꿀 인덱스 리스트`

```python
df1 = df.copy() # 원본 안덱스를 보호하려고 하는거임
df1.rename({0:'row1', 1:'row2'}) # 0에 해덩하는 인덱스를 row1로 1에 해당하는 인덱스를 row2로 바꿔
```

```python
 df1.head()
```

```python
df1 = df1.rename({0:'row1', 1:'row2'})
df1.head()
```

```python
df1.index = [i+1 for i in range(len(df1))] #인덱스 전체를 바꾸는거
df1.head()
```

### **열을 인덱스로 설정**
- `데이터명.set_index(컬럼명)`

```python
df1.set_index('PassengerId') # 특정 컬럼을 인덱스로 설정 passengerid가 인덱스로 간걸 볼 수 있음
```

```python
df1.head()
```

```python
df1 = df1.set_index('PassengerId') #마찬가지로 변경을 하려면 덧씌워주어야 합니다.
df1.head()
```

### **인덱스를 열로 변환**
- 인덱스를 열로 변환하고 그 열을 남기고 싶으면
    - `데이터명.reset_index()`
- 인덱스를 열로 변환하고 그 열을 삭제하고 싶으면
    - `데이터명.reset_index(drop = True)`

```python
df1.head()
```

```python
df1.reset_index()
```

```python
df1.reset_index(drop=True) # reset drop true하니까 인덱스 컬럼 없어지고 기존 인덱스가 됨
```

```python
df1 = df1.reset_index() #다시 데이터를 저장해줘야 됨
df1.head()
```

## **[2] 행**

### **행의 추가/제거**
- 행의 추가
    - `pd.concat([기존 데이터명, 붙일 데이터명])`
- 행의 제거
    - `데이터명.drop(인덱스명, axis=0)`

```python
df1 = df.copy()
df2 = df.copy()

df1
```

```python
df2
```

```python
df3 = pd.concat([df1, df2]) #df1과 df2를 리스트로 묵어서 합쳐줌
df3
```

```python
df3 = df3.reset_index(drop=True)
df3 # 중복된 데이터
```

```python
df3.drop([i for i in range(891, len(df3))])
```

### 행 중복 제거
- `데이터명.drop_duplicates()`

```python
df3 # 기존에 중복된 데이터
```

```python
df3.drop_duplicates() # 중복된거 제거해서 원래대로 온거
#왜? 똑같은걸 두개 만들어서 합친거니까
```

## **[3] 열**

### **열의 추가/제거**

- 열의 추가
    - `데이터명[추가할 컬럼명] = 추가할 값`
- 열의 제거
    - `데이터명.drop(제거할 컬럼명, axis=1)`


```python
df.head()
```

```python
df1 = df.copy()
df1['age_simplified'] = df1['Age']//10 * 10 # 열을 추가한고
```

```python
df1.head() # 그래서 20대인지 30대인지 확인할 수 있게 됨
```

```python
print(list(df1['Name'].str.split(',')))
#문자열 컬럼에 str.split()을 쓰면 특정 문자열을 기준으로 자를 수 있습니다.
```

```python
df1['given_name'] = [i[0] for i in df1['Name'].str.split(',')]
df1.head()
```

```python
df1['string_add'] = df1['Sex']+df1['given_name']
df1.head()
```

```python
df1['number_add'] = df1['Age'] + df1['age_simplified']
df1.head()
```

```python
df1.drop('given_name', axis=1) #given_name이 삭제된걸 볼 수 있음
```

### **열 이름 변경**
- 열 이름 하나를 바꿀 때
    - `데이터명.rename({열이름:바꿀이름, 열이름:바꿀이름, ...}, axis=1)`
- 열 이름 전체를 바꿀 때
    - `데이터명.columns = 열 이름 리스트`

```python
df1 = df.copy() # 기존 데이터를 보존하기 위해 일단 복사하고
df1.head()
```

```python
df1.rename({'PassengerId':'Id'}, axis=1) # ID 바뀐거 보이지
```

```python
df1.columns = [i for i in range(12)] # 통째로 열을 바꾸고 싶으면 columns = 해놓고 바꾸면 됨
df1.head()
```
