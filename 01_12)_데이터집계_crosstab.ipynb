# **crosstab**
- 범주형 데이터를 비교분석할 때 유용합니다.
- `pd.crosstab(index=행, columns=열, margins=True/False, normalize=True/False)`

```python
import pandas as pd

file_path = '/content/drive/MyDrive/data/titanic_train.csv'
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

## **범주별 갯수 구하기**
- `pd.crosstab(행, 열)`

```python
pd.crosstab(df['Sex'], df['Survived'])
#성별을 기준으로 생존율을 구할 수 있음
```

```python
pd.crosstab(df['Pclass'], df['Survived'])
#pclass별로 survived가 몇개인지 알 수 있음
```

## **범주별 비율 구하기**
- `normalize = 'all'`: 전체 합이 100%
- `normalize = 'index'`: 행별 합이 100%
- `normalize = 'columns'`: 열별 합이 100%

```python
pd.crosstab(df['Sex'], df['Survived'], normalize='all')
#비율을 구할때 normalize = all
# 전체를 100으로 봤을때 각각에 해당하는 비율을 나타내는거
# 여자고 생존이면 9퍼
# 남자고 생존은 5퍼
```

```python
pd.crosstab(df['Sex'], df['Survived'], normalize='index')
# normalize = index이면 행별로 인자를 계산하기 때문에
# 여자중에서 0인경우 25퍼 이렇게 구할 수 있는거
```

```python
pd.crosstab(df['Sex'], df['Survived'], normalize='columns')
#컬럼의 합이 100인경우로 나타남
```

```python
pd.crosstab(df['Sex'], df['Survived'], normalize='all', margins=True)
#margins = True하면 행과열의 합을 구할 수 있음
```

```python
pd.crosstab(df['Sex'], df['Survived'], normalize='index', margins=True)
```

```python
pd.crosstab(df['Sex'], df['Survived'], normalize='columns', margins=True)
```

## **다중 인덱스, 다중 컬럼의 범주표 구하기**

```python
pd.crosstab(index=[df['Sex'], df['Pclass']], columns=df['Survived'])
#인덱스를 여러개 적고
# survived를 컬럼에 놓았을 떄
# 각각 행의 개수가 몇개인지 나오는거임
```

```python
pd.crosstab(index=[df['Sex'], df['Pclass']],
            columns=df['Survived'], normalize='all')
```

```python
pd.crosstab(index=[df['Sex'], df['Pclass']], columns=df['Survived'],
            normalize='columns')
```

```python
pd.crosstab(index=[df['Sex'], df['Pclass']], columns=[df['Survived'],
                                                      df['Embarked']])
```

```python
pd.crosstab(index=[df['Sex'], df['Pclass']],
            columns=[df['Survived'], df['Embarked']], normalize='all')
```
