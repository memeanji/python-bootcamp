# **groupby**
- 같은 값을 한 그룹으로 묶어서 여러 가지 연산 및 통계를 구할 수 있습니다.
- `데이터.groupby(컬럼명).연산및통계함수`

```python
import pandas as pd

file_path = '/content/drive/MyDrive/data/titanic_train.csv'
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

```python
df['PassengerId'] = df['PassengerId'].astype(str)
#passengerid는 문자열로 취급해서 계산을 하지 않을거기때문에
#문자열로 타입변환을 한거임
```

## **단일 그룹**
|함수|설명|
|:--|:--|
|count()|행의 갯수|
|nunique()|행의 유니크한 갯수|
|sum()|합|
|mean()|평균|
|min()|최솟값|
|max()|최댓값|
|std()|표준편차|
|var()|분산|

```python
df.groupby('Pclass').count()
#pclass로 묶어서 행의 개수를 알려달라는
```

```python
df.groupby('Pclass').nunique() #유니크한 갯수
#survived는 생존 또는 사망으로 유니크한 개수 2로 나오게 되는거임
```

```python
df.groupby('Pclass').sum(numeric_only=True)
```

```python
df.groupby('Pclass').mean(numeric_only=True)
```

```python
df.groupby('Pclass').min(numeric_only=True)
```

```python
df.groupby('Pclass').max(numeric_only=True)
```

```python
df.groupby('Pclass').std(numeric_only=True)
```

```python
df.groupby('Pclass').var(numeric_only=True)
```

```python
df.groupby('Pclass')[['Survived']].mean()
#pclass로 묶어서 survived 열만 가져와서 평균값 계산한거임
```

```python
df.groupby('Pclass')[['Survived','Age']].mean()
# [ ] ,대괄호로 묶이면 리스트니까
# 여러개를 가져오고 싶으면 묶고싶은(=기준이되는) 컬럼을 () 안에 넣고
# 대괄호안에 리스트로 묶어서 가져오면 됨
```

## **다중 그룹**

```python
df.groupby(['Sex','Pclass']).mean(numeric_only=True)
#Sex와Pclass 기준으로 평균을 계산한거임
# 남자이고 pclass가 3인 경우 생존율이 가장 낮았구나
# 여자이고 1등석에 앉은 사람들의 생존율이 가장 높았구나
```

```python
import numpy as np

df.groupby(['Sex','Pclass'])[['Survived','Age','SibSp','Parch','Fare']].aggregate([np.mean, np.min, np.max])

#aggregate를 하면 이렇게 그룹바이별로 각 열의 통계값을 볼 수 있음
```
