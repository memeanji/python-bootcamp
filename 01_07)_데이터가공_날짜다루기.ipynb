## **문자형을 날짜형으로 변경**
- 날짜가 문자형으로 되어있다면 날짜형으로 변경해야 여러가지 날짜 계산을 할 수 있습니다.
- `pd.to_datetime(컬럼, format='날짜 형식')`

|형식|설명|
|:--|:--|
|%Y|0을 채운 4자리 연도|
|%y|0을 채운 2자리 연도|
|%m|0을 채운 월|
|%d|0을 채운 일|
|%H|0을 채운 시간|
|%M|0을 채운 분|
|%S|0을 채운 초|

```python
# 날짜를 변경하고 싶으면 pd.to_datatime
pd.to_datetime(df['Date'], format='%Y-%m-%d')
#년월일이 - 을 통해서 연결어 있음
```

```python
df['Date1'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.info()

#format='%Y-%m-%d'이렇게 변환한 데이터를 새로운 컬럼에 넣어보고
# 기존컬럼은 object라는  dtype을 가지고 있지만
#새로운 컬럼은 datatime이라는 dtype을 가지고 있음
```

```python
df.head()
```

## **날짜를 원하는 형식으로 변경**
`데이터컬럼.dt.strftime(날짜형식)`

```python
df['Date1'].dt.strftime('%Y-%m')

#위에처럼 쓰면 년 월까지만 나오고
```

```python
df['Date1'].dt.strftime('%m-%d %H:%M')

# 시간분초가 들어있지 않아써 기본적으로 00시 00분으로 나오는걸 확인가능
```

## **dt 연산자**
|연산자|설명|
|:--|:--|
|year|연도|
|month|월|
|day|일|
|dayofweek|요일(0-월요일, 6-일요일)|
|day_name()|요일을 문자열로|

```python
df['year'] = df['Date1'].dt.year
df['month'] = df['Date1'].dt.month
df['day'] = df['Date1'].dt.day
df['dayofweek'] = df['Date1'].dt.dayofweek
df['dayname'] = df['Date1'].dt.day_name()
```

```python
df.head()
```

## **날짜 계산**
- day 연산: `pd.Timedelta(day=숫자)`
- month 연산: `DateOffset(months=숫자)`
- year 연산: `DateOffset(years=숫자)`

```python
df['plus day1'] = df['Date1'] + pd.Timedelta(days=1)
df.head()
#pandas의 Timedelta 함수를 사용하면 하루전후 한달전후 등 자유롭게 사용가능
```

```python
df['plus day7'] = df['Date1'] + pd.Timedelta(days=7)
df.head()
#timedelta는 일자만 가능
```

```python
df['minus day7'] = df['Date1'] - pd.Timedelta(days=7)
df.head()
```

```python
from pandas.tseries.offsets import DateOffset

df['plus month1'] = df['Date1'] + DateOffset(months=1)
df.head()
#한달을 더해줌
```

```python
df['minus month3'] = df['Date1'] - DateOffset(months=3)
df.head()
```

```python
df['plus year1'] = df['Date1'] + DateOffset(years=1)
df.head()
```

```python
df['minus year3'] = df['Date1'] - DateOffset(years=3)
df.head()
```

## **날짜 구간 데이터 만들기**
`pd.date_range(start=시작일자, end=종료일자, periods=기간수, freq=주기)`

|형식|설명|
|:--|:--|
|D|일별|
|W|주별|
|M|월별 말일|
|MS|월별 시작일|
|A|연도별 말일|
|AS|연도별 시작일|

```python
pd.date_range(start='2020-01-01', periods=30, freq='D')
#하루씩 구간으로 날짜 구간을 구할 수 있음
```

```python
pd.date_range(start='2020-01-01', end='2023-06-30', freq='M')
#freq='M' 월단위로 나옴
#기본적으로 월 초랑 월말의 날짜가 나오는데
```

```python
pd.date_range(start='2020-01-01', end='2023-06-30', freq='MS')
#freq='MS' 이렇게 하면 각 1일만 나옴
```

```python
pd.date_range(start='2020-01-01', end='2023-06-30', freq='A')
#freq='A' 연단위로 나오고 연말과 연초의 날짜가 나오는데
```

```python
pd.date_range(start='2020-01-01', end='2023-06-30', freq='AS')
# 연기준 1일날짜를 뽑고 싶을떄 AS
```

## **기간 이동 계산**
`컬럼.rolling().집계함수`

```python
df1 = data.copy()
df1.head()
```

```python
#7일 이동평균
df1['ma7'] = df1['Temp'].rolling(7).mean()
df1.head(20)
#.rolling(7).mean() 7일 이동평균을 구하는거임
#
```

```python
df1['ma30'] = df1['Temp'].rolling(30).mean()
df1.head(30)
#30일 이동평균
```

- 평균 외에 합계, 최솟값, 최댓값 등 다양한 연산이 가능합니다.

```python
df1['Temp'].rolling(7).sum()
#이동하면서 구간별 연산을 해야할
```

```python
df1['Temp'].rolling(7).min()
```

```python
df1['Temp'].rolling(7).max()
```

## **행 이동**
`컬럼.shift(이동할 행의 수)`

```python
df2 = data.copy()
```

```python
df2.head()
```

```python
df2['Temp shift1'] = df2['Temp'].shift(1)
df2.head()
#shift(1) 하면 한행씩 밀린걸 확인할 수 있음
```

```python
df2['pct change'] = (df2['Temp shift1'] - df2['Temp'])/df2['Temp']
df2.head()
#ex. 전날대비 증감율이 얼마나 됐냐 이런걸 확인하려고할 땐
# (df2['Temp shift1'] - df2['Temp'])/df2['Temp'] 이렇게 해주면 됨
```

```python
df2['Temp'].shift(7).head(10)
#이렇게 7을 넣게 되면 이렇게 7칸이 밀리게 되고
```

```python
df2['Temp'].shift(-1).head(10)
#양수뿐만 아니라 음수도 넣을 수 있음
# 이렇게 앞,뒤로 땡길 수 있음
```
