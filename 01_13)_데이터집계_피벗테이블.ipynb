- `pd.pivot_table(데이터명, index=, columns=, values=, aggfunc=)`

```python
import pandas as pd

file_path = '/content/drive/MyDrive/data/titanic_train.csv'
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

## **단일 인덱스, 단일 컬럼, 단일 값**

```python
pd.pivot_table(df, index='Sex', columns='Pclass',
               values='Survived', aggfunc='mean')

# 피벗테이블은 pd.pivot_table이라는 함수 안에다가 데이터를 먼저 넣고
# index, col,values에 들어갈 컬럼이름을 적어놓고 aggfunc 연산할 함수를 넣으면 됨
# 여자이고 객실등급이 높을 수록 생존율이 높았음
```

```python
pd.pivot_table(df, index='Pclass', columns='Sex', values='Survived',
               aggfunc='mean')
```

```python
pd.pivot_table(df, index='Sex', columns='SibSp', values='Survived',
               aggfunc='mean')

# 동반가족이 없을수록 생존확률이 높았다라고 나옴
```

```python
pd.pivot_table(df, index='Sex', columns='Survived', values='Age', aggfunc='mean')

#성별에 따른 생존율의 나이의 평균을 구한거
```

- `margins` 옵션을 통해 행과 열 전체의 값도 구할 수 있습니다.

```python
pd.pivot_table(df, index='Sex', columns='Survived', values='Age',
               aggfunc='mean', margins=True)

# margins = True라는 인자를 넣어주면 행과열 전체의 값도 구할 수 있음
```

```python
pd.pivot_table(df, index='Sex', columns='Survived', values='Age',
               aggfunc=['max','min','mean'])

# 여러개의 연산을 하고 싶으면 aggfuc에 여러개 넣으면 최대 최소 평균 모두 구할 수 있음
```

## **다중 인덱스, 다중 컬럼, 다중 값**

```python
pd.pivot_table(df, index=['Sex','Pclass'], columns='Survived', values='Age',
               aggfunc='mean')

# 여러개를 넣고 싶으면 이렇게 대괄호로 감싸는 리스트의 형태로 주면 됨
```

```python
pd.pivot_table(df, index='Survived', columns=['Sex','Pclass'], values='Age',
               aggfunc='mean')
```

```python
pd.pivot_table(df, index=['Sex','Pclass'], columns=['Survived','Embarked'],
               values='Age', aggfunc='mean')
```

```python
pd.pivot_table(df, index=['Sex','Pclass'], columns=['Survived','Embarked'],
               values=['Age','Fare'], aggfunc='mean')
```
