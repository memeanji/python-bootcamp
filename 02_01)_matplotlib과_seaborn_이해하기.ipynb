# **matplotlib과 seaborn 이해하기**


---



- **`matplotlib`**: 파이썬 데이터 시각화의 가장 기본적인 라이브러리. 유연하게 커스텀이 가능하나 코드가 길고 어렵다는 단점이 있습니다.
- **`seaborn`**: **`matplotlib`**을 좀 더 쉽고 아름답게 시각화하는 라이브러리
- 실제 분석에서는 **`seaborn`**과 **`matplotlib`**을 함께 사용하여 시각화하는 경우가 많습니다.

## **[1] 기본 문법**
```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(가로길이, 세로길이)) #그래프 크기 설정
sns.000plot(data=데이터, x=X축 컬럼, y=Y축 컬럼, hue=범례 컬럼) #seaborn으로 그래프 그리기
plt.title(제목) #제목 설정
plt.xlabel(라벨) #x축 라벨 설정
plt.ylabel(라벨) #y축 라벨 설정
plt.legend(loc=범례 위치 설정)
plt.xticks(rotation=x축 각도 설정)
plt.yticks(rotation=y축 각도 설정)
plt.show() #그래프 출력
```



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = sns.load_dataset('penguins')
df
````

````python
#Bar그래프
sns.barplot(data=df, x='species', y='body_mass_g')
#종별로 몸무게 평균을 나타내줌
````

````python
plt.figure(figsize=(10,5)) #크기
sns.barplot(data=df, x='species', y='body_mass_g') # x,y축
plt.title('body mass per species') # 제목
plt.xlabel('per species') # 기본적으로 컬럼명이 들어가는데
plt.ylabel('body mass') # 원하는게 있으면 넣어줌
plt.show()
````

````python
df.head()
````

````python
#성별로 나눠서 Barplot
sns.barplot(data=df, x="island", y="body_mass_g", hue="sex")
plt.show()
#x축 섬이름 y축 몸무게 평균
#암컷이 몸무게가 적고
#Bisco 섬의 펭귄들의 몸무게가 높음
````

````python
sns.barplot(data=df, x="island", y="body_mass_g", hue="sex")
plt.xticks(rotation=30) #x축의 글자가 30도 회전
# x축에 글자가 많이들어갈때 기울이면 더 잘보임
plt.legend(loc='upper left') # 범례의 위치를 바꿀 수 있음
plt.show()
````

## **[2] 스타일 설정하기**

```
sns.set_style(스타일)
sns.set_palette(팔레트)
```
- 스타일: darkgrid, whitegrid, dark, white
- 팔레트: [참고](https://seaborn.pydata.org/tutorial/color_palettes.html)


````python
sns.set_style('darkgrid')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('whitegrid')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('white')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('dark')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_palette('Set3')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('white')
sns.set_palette('flare')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_palette('crest')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````
