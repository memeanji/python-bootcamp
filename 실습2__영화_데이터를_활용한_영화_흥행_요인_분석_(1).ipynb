## **데이터 전처리**

- 필요한 컬럼만 남기기

```python
movies_df = movies[['id','budget','genres','title','release_date','revenue','vote_average','vote_count']]
credits_df = credits[['movie_id','crew','cast']]
```

- 데이터 결합

```python
data = pd.merge(movies_df, credits_df, left_on = 'id', right_on = 'movie_id')
data.head()
#지금 id와 movie id 값이 같은데 두개로 나눠져있어서 밑에는 하나 삭제한거임
```

```python
data = pd.merge(movies_df, credits_df, left_on = 'id', right_on = 'movie_id').drop('movie_id', axis=1)
data.head()
```

- roi 컬럼 만들기

```python
data['roi'] = data['revenue'] / data['budget'] # 수익 / 예산 = roi
data.head()
```

- 감독 컬럼 만들기

```python
data['crew'][0]
```

```python
import ast

print(ast.literal_eval(data['crew'][0]))
```

```python
data['crew'] = data['crew'].apply(ast.literal_eval)
```

```python
print(data['crew'][1])
```

```python
def get_director(x): # 요소가 director일 경우에 name을 추출한다라는 함수를 만들어주는거임
    for i in x: # list안에 요소들을 돌면서
        if i['job'] == 'Director': #만약 key값이 job이고 value가 Director라면
            return i['name'] # name이라는 값을 return하라
```

```python
get_director(data['crew'][0])
```

```python
data['director'] = data['crew'].apply(get_director)
```

```python
data.head() # 감독컬럼 생성까지 완료
```

- 배우 컬럼 만들기

```python
data['cast'][0] # 리스트로 들어가있고 딕셔너리로 묶어놨구나
```

```python
[i['name'] for i in  ast.literal_eval(data['cast'][0])] # 이걸 람다 함수를 통해서 넣는거임
```

```python
data['cast_name'] = data['cast'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])
data.head()
```

- 장르 컬럼 만들기

```python
data['genres'][0]
```

```python
data['genres'] = data['genres'].apply(ast.literal_eval)
```

```python
def get_genres(x):
    if len(x) > 0 :
        return x[0]['name']
```

```python
get_genres(data['genres'][0])
```

```python
def get_genres(x):
    if len(x) > 0: # 리스트안에 뭐가 있으면
        return x[0]['name'] # 첫번째 인덱스에서 name이라는 key를 찾아서 value를 가져와라라는 뜻

data['main_genre'] = data['genres'].apply(get_genres) # main 장르라는 컬럼을 생성시킨거
data.head()
```

- 데이터 타입 변경

```python
data.info() # 기존의 데이터 타입을 살펴보는거

#id값은 유니크한 값으로 str으로 바꿔줄거임
# release date는 날짜 형식이기 때문에 날짜로 바꿔줄거임
```

```python
#pandas에서 날짜로 바꿔주는 함수 to_datatime
data['release_date'] = pd.to_datetime(data['release_date'], format='%Y-%m-%d') #년-월-일이였으니까 포맷을 맞춰서
data['id'] = data['id'].astype(str) #id는 astype으로 str으로 바꾸는거
```

```python
data.info()
```

- 연도, 월 컬럼 만들기

```python
data['year'] = data['release_date'].dt.year
data['month'] = data['release_date'].dt.month
data.head()
```

- 결측치 제거



```python
data.info()
```

- 예산 TOP 10, 투표수 TOP 10 영화는?

```python
title_dic = {'budget':'예산', 'vote_count':'투표수'}
for y in ['budget','vote_count']:
    top = data.groupby('title')[[y]].sum().reset_index().sort_values(y, ascending=False).head(10)
    fig = px.bar(data_frame=top, x='title', y=y, title=f"{title_dic[y]} TOP 10 영화",
                 color_discrete_sequence=['#AFCBFF'] )
    fig.update_layout(
        paper_bgcolor='white',
        plot_bgcolor='white',
        yaxis=dict(
            showgrid=True,
            gridcolor='lightgray'
        )
    )
    fig.show()  # 각 반복마다 출력

    #예산수익이 높으면 흥행도 높을거라고 생각했는데 그건 아니구나
    #투표수 높은걸 보니까 인셉
```



> 대체적으로 예산, 투표수가 높으면 흥행 수익도 높아보이나, 정비례하진 않아 보인다. -> 상관관계 분석을 진행해보자.

> 액션과 드라마 장르에 흥행 수익이 아주 높은 영화들이 많이 있지만, 중앙값을 비교하면 다른 장르에 비해 높은 편은 아니다.

> 장르별 영화 수익 평균은 애니메이션 > 어드벤쳐 > 가족 > SF > 판타지 > 액션 순이다.

> 흥행 수익 합계는 액션 > 어드벤쳐 > 드라마 > 코미디 > 애니메이션 순이다.






> 2010년도부터 액션 영화의 흥행 수익 비중이 높아지기 시작했다.

> 1990년대 후반 ~ 2000년대 초반에는 드라마 장르의 흥행 수익이 높았다.



> 액션, 어드벤처 장르는 비교적 봄, 여름에 개봉 수익이 높았다.

> 드라마 장르는 비교적 가을, 겨울에 개봉 수익이 높았다.

> 코미디 장르는 비교적 겨울에 개봉 수익이 높았다.




> 예산과 투표수는 흥행 수익과 높은 양의 상관관계를 보이나, 평점 평균은 비교적 낮은 양의 상관관계를 보인다.

> 흥행에 성공한 영화 상위 100개로 추려보면, 그 상관관계가 더 낮아진다.






> 액션의 경우 예산이 큰 영화가 많아 ROI가 높은 편은 아니고, 드라마/코미디/로맨스 영화가 ROI가 큰 영화들이 많다.
