## **apply 함수**
- 사용자 정의 함수를 데이터에 적용하고 싶을 때 사용합니다.
- `.apply(함수, axis=0/1)`

```python
#apply함수가 데이터를 가공할때 갱장히 유용하기 때문에 기억하기
df1 = df.copy()
```

```python
def pclass_sibsp(x): #사용자정의함수를 만들어 놓은거
    if x['Pclass'] == 1 and x['SibSp'] == 1:
        return 1
    else:
        return 0
```

```python
df1['pclass_sibsp_filter'] = df1.apply(pclass_sibsp, axis=1)
#열의 이름을 참조헀기 때문에 axis = 1을 꼭 적어주기
```

```python
df1.head(5)
#Sibsp는 동반자이고
# 1등석이고 동반자가 있으면 1 , 동반자가 없으면 0
#새로운 pclass_sibsp_filter 열이 생긴걸 확인할 수 있음
```

```python
import numpy as np

def adult(x):
    if x >= 19:
        return 1
    elif x < 19:
        return 0
    else:
        return np.nan
```

```python
df1['adult_yn'] = df1['Age'].apply(adult)
df1.head(5)
#adult_yn 새로운 열이 생김
```

- 간단한 함수는 lambda로 구현할 수 있습니다.

```python
df1['pclass_sibsp_lambda'] = df1.apply(lambda x: 1 if x['Pclass'] == 1 and x['SibSp'] == 1 else 0, axis=1)
df1.head(3)
# 람다함수를 사용해서 확인할 수 있음
```

## **map 함수**
- 값을 특정 값으로 치환하고 싶을 때 사용합니다.
- `데이터명[컬럼명].map(매핑 딕셔너리)`

```python
gender_map = {'male':'남자', 'female':'여자'} # 딕셔너리를 이용해서
df1['Sex_kr'] = df1['Sex'].map(gender_map)
df1.head(3)
#map함수는 특정한 값을 다른 값으로 치환하고 싶을 떄 사용
```

## **문자열 다루기**
|메소드|설명|
|:--|:--|
|`.str.contains(문자열)`|문자열을 포함하고 있는지 유무|
|`.str.replace(기존문자열, 대치문자열)`|문자열 대치|
|`.str.split(문자열, expand=True/False, n=개수)`|특정 문자열을 기준으로 쪼개기|
|`.str.lower()`|소문자로 바꾸기|
|`.str.upper()`|대문자로 바꾸기|

```python
df2 = df.copy()
df2.head(3)
```

- `.str.contains(문자열)`

```python
df2['Name'].str.contains('Mrs') # T면 포함 F는 미포함
#특정문자열을 포함하고 있는지 contains()
```

```python
df2[df2['Name'].str.contains('Mrs')]
# 이조건을 대괄호 안에 넣게 되면 이 문자열을 포함한 열을 추출할 수 있음
```

- `.str.replace(기존문자열, 대치문자열)`

```python
df2['Name'] = df2['Name'].str.replace(',', '*** ')
df2.head(10)
#기존의 문자열을 다른 문자열로 대치 시켜주는
#콤마가 모두 사라진걸 확인할 수있음
```

- `.str.split(문자열, expand=True/False, n=개수)`

```python
df2['Name'].str.split(' ')
#특정 문자열 기준으로 쪼개기 가능
#빈칸 기준으로 쪼개진거임
```

```python
df2['Name'].str.split(' ', expand=True, n=2)
#expand = False가 디폴트 값이고 True라는 값을 넣게 되면
# 열을 쪼개서 펼쳐주게 됨
# expand를 쓸때 몇개의 열로 펼쳐줄건지
# n = 1이니까 한개로 펼쳐달라고 한거임
```

- `.str.lower()`
- `.str.upper()`

```python
df2['Name'].str.lower()
#소문자 변환
```

```python
df2['Name'].str.upper()
#대문자 변환
```
